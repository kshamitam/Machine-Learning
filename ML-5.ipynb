```python
trans=[]
for i in range(0,7501):
    trans.append([str(data.values[i,j])for j in range(0,20)])
    
trans
```

```python
pip install apyori
```

```python
from apyori import apriori
rules= apriori(trans, min_support=0.05, min_confidence=0.5, min_lift=0.5, min_length=2)
rules
```

```python
final = list(rules)
final
```

```python
for i in final:
    print(i)
    print("/n")
```

```python
np= pd.DataFrame(final)
np = np.drop(columns=['ordered_statistics'])
np = np.drop(columns=['support'])

np
```

```python
fs= pd.DataFrame(final, columns=["support","confidence","lift"] )
fs = fs.drop(columns=['support'])
fs = fs.drop(columns=['lift'])
fs
```

```python
result = pd.concat([np, fs], axis=1, join='inner')
result
```

```python
df_sorted = result.sort_values(by='confidence', ascending=False)
df_sorted
```
